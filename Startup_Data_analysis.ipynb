```python
import numpy as np
import pandas as pd
import seaborn as sns
import warnings as wn
import matplotlib as plt

pd.set_option('mode.chained_assignment',None)
pd.set_option('display.max_colwidth', -1)
pd.options.display.max_columns = 50
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
%matplotlib inline
sns.set(rc={'figure.figsize':(50,30)})
```

```python
dir_path = "C:/Users/vik2nu/Desktop/AI_ML/Term1_2_Project/01_Startup_Data/"
sfinc=pd.read_excel(dir_path+'Startup_information_cleaned.xlsx')
sfinc.info()
```

```python
sns.set(rc={'figure.figsize':(50,30)})
a=sns.countplot('Sorted_Cities', hue='sort_ind_vert',data = sfinc)
a.tick_params(labelsize=30)
a.set_yscale("log")
plt.legend(loc='upper right')
plt.setp(a.get_legend().get_texts(), fontsize='28')
a.set_ylabel("No. of investments",fontsize=40)
```

```python
sfinc['Sorted_Cities'].value_counts().head(8).plot(kind='bar', figsize=(10,5),colormap='rainbow',fontsize=13, xticks=np.arange(0,15,2))
```

```python
space = np.ones(7)/28
sfinc['Sorted_Cities'].value_counts().plot(kind='pie',explode=space,fontsize=14, autopct='%.2f%%',figsize=(10,10),cmap='inferno',wedgeprops=dict(width=0.15),legend=True,startangle=200, shadow=True )
```

```python
c=sns.countplot('Sorted_Cities', hue='invst_type_sorted',data = sfinc)
c.tick_params(labelsize=30)
plt.legend(loc='upper right')
plt.setp(c.get_legend().get_texts(), fontsize='40')
c.set_yscale("log")
c.set_ylabel("No. of investments",fontsize=40)
```

```python
d=sns.boxplot(data=sfinc, x='invst_type_sorted', y='amount',palette='rainbow',showfliers=False)
d.tick_params(labelsize=40)
d.set_yscale("log")
d.set_ylabel("Amount Invested $",fontsize=40)
```
